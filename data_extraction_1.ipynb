```python
import tweepy, csv
import pandas as pd
```

```python
consumer_key = 'xx'
consumer_secret = 'xx'
access_token = 'xx-'
access_token_secret = 'xx'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
tweets = {}

for tweet in tweepy.Cursor(api.search,q="#iphonexs",count=30000, lang="en", since="2018-04-03", tweet_mode="extended").items():
    tweets[tweet.created_at] = tweet.full_text
```

```python
new_dict = {'Date':[], 'Post':[]}
for i in tweets:
    new_dict['Date'].append(i)
    new_dict['Post'].append(tweets[i])
```

```python
df = pd.DataFrame.from_dict(new_dict)
```

```python
df = df[~df['Post'].str.contains("giving away")]
df = df[~df['Post'].str.contains("giveaway")]
df = df[~df['Post'].str.contains("Giveaway")]
df = df[~df['Post'].str.contains("ANkB3J")]


df.index = range(len(df.index))
df
```
